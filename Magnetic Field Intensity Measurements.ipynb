```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 0, 0, 0, 0, 0, 0, 2 , 2, -2, -2, -2, 2, 0, 0])
y = np.array([-1.75, 0, 1.75, -1.75, 1.75, -1.75, 1.75, -1.75, 1.75, -1.75, 1.75, 0, 0, 0, 0])
z = np.array([0, 0, 0, 2, 2, -2, -2, 0, 0, 0, 0, 0, 0, 2, -2])
B = np.array([0.124, 0.108, 0.122, 0.264, 0.240, 0.294, 0.324, 0.362, 0.272, 0.252, 0.270, 0.092, 0.089, 
              0.089, 0.090])
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel("X-axis", fontsize=15,)
ax.set_ylabel("Y-Axis", fontsize=15,)
ax.set_zlabel("Z-Axis", fontsize=15,)
plt.title("Magnetic Field Intensity Between the Pole Faces", fontsize=16)

img = ax.scatter(x, y, z, c=B, cmap=plt.winter())
fig.colorbar(img)
ax.view_init(10, 20)
plt.show()
```
