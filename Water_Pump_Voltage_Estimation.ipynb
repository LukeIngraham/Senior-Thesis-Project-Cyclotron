```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
volt = np.array([1, 2, 3, 4, 5])
Volume = np.array([0, 360, 555, 720, 850])

def fit(volt, Volume):
    # calculate polynomial
    z = np.polyfit(Volume, volt, 3)
    f = np.poly1d(z)

    # calculate new x's and y's
    x_new = np.linspace(Volume[0], Volume[-1], 50)
    y_new = f(x_new)

    plt.figure(figsize=(14,8))
    plt.title("Fit of the water pumped with a 3 order polynomial", size=15)
    plt.xlabel("Volume [mL]", size=12)
    plt.ylabel("Voltage [V]", size=12)
    plt.plot(Volume, volt, 'tab:blue',label = 'Curve')
    plt.plot(Volume,volt,'ko', markersize = 8,  label = 'peaks')
    plt.plot( x_new, y_new, 'r',label ='Fit_order 3')
    plt.legend(fontsize=14)
    plt.grid()
    plt.show()
    print("Each measurment taken after 30 seconds.")
    print("Slope =")
    print(f)
```

```python
#multiplying the volume measurments by 2 to account for the 1 minute flow rate requirment.
Volume2 = np.array([0, 360, 555, 720, 850])*2
fit(volt, Volume2)
```
