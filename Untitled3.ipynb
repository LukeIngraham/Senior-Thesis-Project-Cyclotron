```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def number_of_ions(I, t):
    e = 1.60*10**(-19)
    return t/(I*e)

def electron_E(temp):
    A_0 = 1.20173e6 #A•m^-2•K^-2
    k = 1.380649e-23 #J/K
    lam_0 = 0.5
    A_g = lam_0 * A_0 #A•m^-2•K^-2
    T = temp + 273.5 # Kelvin
    W = 7.2098e-19 # Joules
    J = A_g*(T**(2))*np.exp(-W/(k*T))
    return J 

# Finds the number of electrons generated
# J is the current density
def electron_num(J):
    electrons = 6.241509e18 #electron/second
    return J*electrons
```

```python
current_data = np.array([0.00017, 0.00023, 0.00026, 0.00028])-0.00004
time_data = np.array([1 ,60, 138, 203])
number_ion = np.zeros(len(current_data))

for i in range(0, len(current_data)):
    number_ion[i] = number_of_ions(current_data[i], time_data[i])
print(number_ion)

plt.figure(figsize=(10,7))
plt.plot(time_data, number_ion)
plt.xlabel("Time [s]", size=12)
plt.ylabel("Number of Electrons", size=12)
plt.title("Time vs. Number of Electrons", size=18)
plt.grid()
plt.show()
```

```python
current_data2 = np.array([0.00017, 0.00030, 0.00055, 0.00120, 0.00250, 0.00451, 0.00700, 0.01101, 0.01440, 0.01776])-0.00004
time_data2 = np.array([1 ,99, 119, 148, 171, 187, 205, 223, 236, 251])
number_ion2 = np.zeros(len(current_data2))

for i in range(0, len(current_data2)):
    number_ion2[i] = number_of_ions(current_data2[i], time_data2[i])
print(number_ion2)

plt.figure(figsize=(10,7))
plt.plot(time_data, number_ion, label="Test 1, 0V")
plt.plot(time_data2, number_ion2, label="Test 2, 10V")
plt.xlabel("Time [s]", size=12)
plt.ylabel("Number of Electrons", size=12)
plt.title("Time vs. Number of Electrons", size=18)
plt.legend(fontsize=12, loc="upper left")
plt.grid()
plt.show()
```

```python
time_projection = np.linspace(0,500,500)
projection_data = np.zeros(len(time_projection))
#print(len(time_projection), len(projection_data))
e = 1.60*10**(-19)
for k in range(0, len(time_projection)-1):
    projection_data[k] = ((electron_num(electron_E(2000))*(3.045*10**(-5)))/e)*time_projection[k]

plt.figure(figsize=(10,7))
plt.plot(time_projection, projection_data,label="Projection")
plt.plot(time_data, number_ion, label="Test 1, 0V")
plt.plot(time_data2, number_ion2, label="Test 2, 10V")
plt.xlabel("Time [s]", size=12)
plt.ylabel("Number of Electrons", size=12)
plt.title("Time vs. Number of Electrons", size=18)
plt.grid()
plt.show()
```
