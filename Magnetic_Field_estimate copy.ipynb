```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def resistance(V, I):
    R = V/I
    return R

def mT_to_T(B):
    new_B = 0.001*B
    return new_B

def Tesla_to_Gauss(B):
    return B*1e4

def Gauss_to_Tesla(B):
    return B/1e4

voltage = np.array([1, 5, 10, 15, 20])
I = np.array([0.236, 1.184, 2.256, 3.533, 4.800])
B = np.array([4.40, 18.5, 30.5, 55.5, 72.8])
B_T = mT_to_T(B)

plt.figure(figsize=(6,4))
plt.plot(voltage, I, label="Data")
m1,b1 = np.polyfit(voltage, I, 1)
I_fit = m1*voltage+b1
plt.plot(voltage, I_fit, label="Fit")
plt.xlabel("Voltage [V]")
plt.ylabel("Current [A]")
plt.legend()
plt.grid()

plt.figure(figsize=(6,4))
m,b = np.polyfit(I, B_T, 1)
plt.plot(I, B_T, label="Data")
B_fit = m*I+b
plt.plot(I, B_fit, label="Fit")
plt.xlabel("Current [A]")
plt.ylabel("B-Field [T]")
plt.legend()
plt.grid()
```
