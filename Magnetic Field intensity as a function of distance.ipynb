```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def B_z(Z):
    MU0 = 4*np.pi*(10**-7) # N/A^2
    R = 5.08 # cm 
    I = 10 # A
    B = (MU0*I*(R**2))/(2*((R**2)+(Z**2))**(3/2))
    return B
```

```python
dist = 9 # cm
intensity = np.linspace(0,dist, dist+100)
```

```python
plt.figure(figsize=(12,8))
plt.plot(intensity, (B_z(intensity)+B_z(dist-intensity))/B_z(0))
plt.xlabel("Gap Distance [cm]", size=12)
plt.ylabel("B-Field Intenisty [T]", size=12)
plt.title("Current vs. B-Field (Normalized)", size=18)
plt.grid()
plt.show()

print(max((B_z(intensity)+B_z(dist-intensity))/B_z(0)))
print(min((B_z(intensity)+B_z(dist-intensity))/B_z(0)))
```
